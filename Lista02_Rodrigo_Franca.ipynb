### Cria uma tabela com DuckDB com os dados do arquivo `healthcare-dataset-stroke-data.csv` que estão no Blackboard

```python
import duckdb
import pandas as pd
```

```python
path = 'healthcare-dataset-stroke-data.csv'

df = pd.read_table(path, sep=',')
df.head()
```

```python
con = duckdb.connect(database='df')
```

### Crie uma query SQL com o número de pessoas com AVC por tipo de trabalho, renomeando a coluna com os valores para `work_type_count`, em ordem decrescente

```python
query = """ 
    SELECT work_type, count(*) as work_type_count
    FROM df
    WHERE stroke = 1
    GROUP BY work_type
    ORDER BY work_type_count DESC
"""

con.execute(query).df()
```

### Crie uma query SQL com o número total e a porcentagem de cada gênero na base

```python
query = """ 
    SELECT gender, count(*) as count_gender,
    ROUND((COUNT(*) * 100.0) / SUM(COUNT(*)) OVER(), 2) as percentual
    FROM df
    GROUP BY gender
"""

con.execute(query).df()
```

### Crie uma query SQL com o número total e a porcentagem de pacientes masculinos com AVC

```python
query = """
    SELECT 
    gender,
    SUM(CASE WHEN gender = 'Male' AND stroke = 1 THEN 1 ELSE 0 END) as count_gender,
    ROUND((SUM(CASE WHEN gender = 'Male' AND stroke = 1 THEN 1 ELSE 0 END) * 100.0) / SUM(CASE WHEN gender = 'Male' THEN 1 ELSE 0 END), 2) as percentage
    FROM df
    WHERE gender = 'Male'
    GROUP BY gender
"""

con.execute(query).df()
```

### Crie uma query SQL com a quantidade de AVCs por faixa etária

```python
query = """
SELECT age, COUNT(*) AS num_stroke
FROM df
WHERE stroke = 1
GROUP BY age
ORDER BY num_stroke DESC;
"""

con.execute(query).df()
```
